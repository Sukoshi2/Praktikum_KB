### Data Pre-Processing

[Music Genre Dataset](https://www.kaggle.com/datasets/vicsuperman/prediction-of-music-genre "Music Genre Dataset")

###### *Import berberapa libary dan pada sklearn kita import yang diperlukan untuk pre-processing*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder


df = pd.read_csv('music_genre.csv') ## dataset yang original langsung dari kaggle
df1 = pd.read_csv('music_genre_rusak.csv') ## dataset yang udah rusak
```

#### Cleaning data Terlebih dahulu (KARENA STANDARISASI SAYA ERROR KARENA BANYAK NILAI NAN/ '?')
- bilangan bulat = median/modus
- bilangan desimal = mean
- Object / tulisan = modus

###### *karena pada dataset dari kaggle orang ngasih nan itu jadi tanda ? maka dari itu sebelum cleaning kita rubah jadi nan/null*

```python

df1.replace("?", np.nan, inplace=True)
```

###### *menggunakan imputer untuk me-replace data yang null menjadi sesuai dengan ketentuan*

```python

imputer_mean = SimpleImputer(strategy='mean') ## mean atau rata-rata
imputer_median = SimpleImputer(strategy='median') #median atau nilai tengah
imputer_most_frequent = SimpleImputer(strategy='most_frequent') # modus

df1['instance_id'] = imputer_mean.fit_transform(df1[['instance_id']])
df1['artist_name'] = imputer_most_frequent.fit_transform(df1[['artist_name']])
df1['track_name'] = imputer_most_frequent.fit_transform(df1[['track_name']])
df1['popularity'] = imputer_mean.fit_transform(df1[['popularity']])
df1['acousticness'] = imputer_mean.fit_transform(df1[['acousticness']])
df1['danceability'] = imputer_mean.fit_transform(df1[['danceability']])
df1['duration_ms'] = imputer_mean.fit_transform(df1[['duration_ms']])
df1['energy'] = imputer_mean.fit_transform(df1[['energy']])
df1['instrumentalness'] = imputer_mean.fit_transform(df1[['instrumentalness']])
df1['key'] = imputer_most_frequent.fit_transform(df1[['key']])
df1['liveness'] = imputer_mean.fit_transform(df1[['liveness']])
df1['loudness'] = imputer_mean.fit_transform(df1[['loudness']])
df1['mode'] = imputer_most_frequent.fit_transform(df1[['mode']])
df1['speechiness'] = imputer_mean.fit_transform(df1[['speechiness']])
df1['tempo'] = imputer_most_frequent.fit_transform(df1[['tempo']])
df1['obtained_date'] = imputer_most_frequent.fit_transform(df1[['obtained_date']])
df1['valence'] = imputer_mean.fit_transform(df1[['valence']])
df1['music_genre'] = imputer_most_frequent.fit_transform(df1[['music_genre']])

df1.isna().sum()
```

#### Buat data copy supaya tidak merusak data asli

```python
df_norm = df1.copy() ## untuk normalisasi
df_std = df1.copy() ## untuk standarisasi
```

#### Membagi (Split) data Train dan test dengan perbandingan 70:30

###### *membagi nilai x sebagai data latih atau train dan y sebagai data uji atau test*

```python
x = df1.iloc[:,:-1]
y = df1.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .3)

print("Jumlah Data X_train : ", x_train.shape)
print("Jumlah Data X_test  : ", x_test.shape)
print("Jumlah Data y_train : ", y_train.shape)
print("Jumlah Data y_text  : ", y_test.shape)
```

#### Melakukan Normalisasi dengan menggunakan MinMaxScaler

###### *normalisasi memanggil min max scaler dan ditransformasi menjadi data yang telah tenormalisasi*

```python

min_max_scaler = MinMaxScaler()
x_scaler = min_max_scaler.fit_transform(df_norm[['popularity','acousticness','danceability',
            'duration_ms', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence']])

df_norm2 = pd.DataFrame(x_scaler)
df_norm2.head(15)
```

#### Mengganti tipe data dari float jadi integer

###### *mengganti tipe data popularity, instance id dari float 64 jadi int 64*

```python

df1['instance_id'] = df1['instance_id'].astype('int64')
df1['popularity'] = df1['popularity'].astype('int64')

df1.info()
```

#### Melakukan one hot encoding

###### *menambah berberapa dari label music genre menjadikan datanya jadi int atau float*

```python
df_noterencoder = df1.copy()

one_hot_encoder = OneHotEncoder(sparse=False)

music_genre_cat = one_hot_encoder.fit_transform(df_noterencoder[['music_genre']])
music_genre_cat = pd.DataFrame(music_genre_cat)

df_terencoder = df_noterencoder.join(music_genre_cat)
df_terencoder.head(10)
```
