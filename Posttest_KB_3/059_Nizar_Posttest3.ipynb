### Visualisasi Data

[Music Genre Dataset](https://www.kaggle.com/datasets/vicsuperman/prediction-of-music-genre "Music Genre Dataset")

##### Pertama Import pandas dan read untuk mengambil data menjadi dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('music_genre.csv')
```

### 1. Visualisasi data menggunakan Grafik Batang / Bar

1. Menggunakan df['namaatribut'] untuk memanggil semua data diddalam atribut itu
2. f = plt.figure() untuk mengatur lebar dan tinggi sebuah visualisasi

#### Hasil Visualisasi :
- menurut diagram batang dibawah, dapat disimpulkan bahwa musik yang ber-genre Anime mempunyai popularitas yang rendah dikarenakan rata-rata elemen masyarakat tidak semuanya tahu tentang bahasa jepang atau animasi jepang
- dan musik yang terpopuler adalah musik yang ber-genre rap yang memiliki popularitas tinggi, hampir semua orang menyukai rap, yaa karena musik yang ber-genre rap itu paling sering populer di berberapa platform misalkan youtube, spotify dll.

```python
## bersihkan data yang NAN/null biar enak melihat visualisasi datanya
df.dropna(inplace=True)

x = df['music_genre'].astype(str)
y = df['popularity']
sakura = '#F999B7'
f = plt.figure()
f.set_figwidth(11)
f.set_figheight(5)
plt.bar(x, y,color=sakura)
plt.title('Musik yang Paling Populer', loc='right')
plt.xlabel('Genre Musik', loc='right')
plt.ylabel('Popularitas', loc='bottom')
plt.show()
```

#### 2. Visualisasi Data Menggunakan Grafik PIE
- mengambil nilai subset dengan cara df.dropna(subset=['key'])
- lalu memanggil pie dari menghitung jumlah subset, dan mengambil jumlah id berdasarkan index

##### Hasil Visualisasi :
- Menurut diagram PIE dibawah, dapat disimpulkan bahwa musik kebanyakan menggunakan kunci G, daripada Kunci D#
- karena Tanda Kunci G digunakan untuk menuliskan nada-nada tinggi misalnya pada jenis Sopran, Mezzo Sopran, Alto. yap seperti diagram diatas yang menyimpulkan bahwa genre rap paling tinggi karena biasanya genre rap menggunakan nada-nada atau suara tinggi
- dan yang paling jarang digunakan adalah tanda kunci D#

```python
dataClean =  df.dropna(subset=['key'])

data_pie = dataClean['key'].value_counts().rename_axis('key').reset_index(name='id_count')

plt.figure(figsize=(10,10))
plt.pie(data_pie.id_count, labels=data_pie.key, startangle=90, autopct='%.1f%%')
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
plt.title('Tanda Kunci Paling Banyak Digunakan')
plt.xlabel('Tanda Kunci Musik', loc='right')
plt.ylabel('Paling Banyak Digunakan', loc='bottom')
plt.show()
```
