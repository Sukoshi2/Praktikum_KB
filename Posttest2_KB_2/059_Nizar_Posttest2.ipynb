```python
import pandas as pd
```

#### IMPORT DATA SET 

```python
df = pd.read_csv('music_genre.csv')
df
```

### Menampilkan data sesuai dengan POSTTEST

1. Jumlah Record
2. Nama Atribut
3. Jumlah Atribut dengan tipe objek, angka, nilai yang NULL

- inisialisasikan data sesuai dengan ketentuan jadi berberapa variabel
- variabel lalu digabung dengan dictionary pada python
- df = importan file csv
- columns = mengambil nama kolom
- len baru digabung dengan columns = menghitung panjang atau banyak data yang berada didalam kolomnya

```python

jumlah_record = len(df)
nama_atrb = df.columns
jumlah_atrb_obj = len(df.select_dtypes(include=['object']).columns)
jumlah_atrb_angka = len(df.select_dtypes(include=['int64', 'float64']).columns)
jumlah_atrb_null = len(df.columns[df.isna().sum() > 0])
```

### Menampilkan Dictionary

menggunakan import series

```python
## menggabungkan semua menjadi satu dictionary untuk dijadikan 1 tabel
dictionary = {
    "Jumlah Record": jumlah_record,
    "Nama Attribute": nama_atrb.to_list(),
    "Jumlah Atribute Objek": jumlah_atrb_obj,
    "Jumlah Atribute Angka": jumlah_atrb_angka,
    "Jumlah Atribute NULL" : jumlah_atrb_null,
}

pd.Series(dictionary)
```

### Menampilkan Nilai 
1. Rata-rata (Mean)
2. Standar Deviasi (Std)
3. Nilai Maksimum (Max)
4. Q1 (Quartil Bawah 25%), Q2 (Quartil Median 50%), Q3 (Quartil Atas 75%)

- menampilkan menggunakan describe untuk menhitung, tapi menggunakan loc untuk mengambil apa saya yang diperlukan dan tidak tampil semua

```python
df.describe().loc[['mean', 'std', 'max', '25%', '50%', '75%']]
```
